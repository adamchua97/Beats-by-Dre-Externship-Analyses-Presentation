```python
import json
import pandas as pd
```

```python
file_path1 = '/content/anker_soundcore_blk_data.json'
with open(file_path1, 'r') as file:
    data1 = json.load(file)
```

```python
reviews_data1 = []
count1 = 0

for result in data1.get('results', []):
    asin = result.get('content', {}).get('asin')
    if asin is None:
        continue  # Skip if asin is missing

    for review in result.get('content', {}).get('reviews', []):
        count1 += 1
        try:
            review_info1 = {
                'review_id': review.get('id'),
                'product_id': asin,
                'title': review.get('title'),
                'author': review.get('author'),
                'rating': review.get('rating'),
                'content': review.get('content'),
                'timestamp': review.get('timestamp'),
                'profile_id': review.get('profile_id'),
                'is_verified': review.get('is_verified', False),
                'helpful_count': review.get('helpful_count', 0),
                'product_attributes': review.get('product_attributes', {}),
            }
            reviews_data1.append(review_info1)
        except KeyError as e:
            print(f"KeyError at review {count1}: {e}")
        except Exception as e:
            print(f"Error at review {count1}: {e}")

if reviews_data1:
    reviews_df1 = pd.DataFrame(reviews_data1)
else:
    print("No reviews data available.")
```
