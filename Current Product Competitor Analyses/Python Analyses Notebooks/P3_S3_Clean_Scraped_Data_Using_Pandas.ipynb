```python
import re

# creating another date column with a 'datetime' data type

# function to extract date information from timestamp string data

def extract_date(review):
  match = re.search(r'(\w+ \d{1,2}, \d{4})', review)
  if match:
    return pd.to_datetime(match.group(1))
  return None

# create new column named 'date' and apply function to 'timestamp'
df['date'] = df['timestamp'].apply(extract_date)


df.head()
```

## Standardize string columns

```python
string_cols = df.select_dtypes(include=['object']).columns
```

```python
# Lower case columns

for col in string_cols:
  df[col] = df[col].str.lower()

df.head()
```
