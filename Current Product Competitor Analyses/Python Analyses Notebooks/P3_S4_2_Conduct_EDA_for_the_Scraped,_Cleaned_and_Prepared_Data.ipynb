## Checking and Handling Outliers

```python
# checking outliers from review length column

Q1 = df['review length'].quantile(0.25)
Q3 = df['review length'].quantile(0.75)
IQR = Q3 - Q1

# create separate dataframe for outliers
outliers = df[(df['review length'] < (Q1 - 1.5 * IQR)) | (df['review length'] > (Q3 + 1.5 * IQR))]

# display outliers information
outliers.info()
```

- Outliers aren’t always bad; they might contain valuable insights, but it’s essential to identify them so you can decide whether to include them in your analysis.
- Outliers will be retained for this dataset to get as much data as we can for the analysis.
